## Insights:
    * Yes outliners has been detected in all items **These outliers value needs to be treated and there are several ways of treating them:
    * Drop the outlier value
    * Replace the outlier value using the IQR

```python
# Define a function which results the upper and lower limit to detect outliers for each fields
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
# Cap & floor the values beyond the outlier boundaries
for i in feature_list:
    LL,UL = remove_outlier(ca_num[i])
    ca_num[i] = np.where (ca_num[i] > UL,UL, ca_num[i])
    ca_num[i] = np.where (ca_num[i] < LL,LL, ca_num[i])
```

```python
# Check the box plot after treating outliers
plt.figure(figsize=(15,15))
feature_list = ca_num.columns
for i in range (len(feature_list)):
    plt.subplot(4, 4, i+1)
    sns.boxplot(y=ca_num[feature_list[i]],data=ca_num)
    plt.title('Boxplot of {}'.format(feature_list[i]))
    plt.tight_layout()
```

```python
# Creating a copy of the original data frame
data_ds = ca.copy()
```

```python
# Copy all the predictor variables into X dataframe
X = data_ds.drop('usr', axis=1)

# Copy target into the y dataframe. 
y = data_ds[['usr']]
```

```python
# To view the top 5 record using head function of the dataset
X.head()
```

```python
# Import the train_test_split library
# Split X and y into training and test set in 70:30 ratio
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30 , random_state=1)
```

# Linear Regression Model

```python
# Import the LinearRegression libary file
from sklearn.linear_model import LinearRegression
```

```python
# invoke the LinearRegression function and find the bestfit model on training data
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
# Let us explore the coefficients for each of the independent attributes
for idx, col_name in enumerate(X_train.columns):
    print("The coefficient for {} is {}".format(col_name, regression_model.coef_[0][idx]))
```

```python
# Let us check the intercept for the model
intercept = regression_model.intercept_[0]
print("The intercept for our model is {}".format(intercept))
```

```python
# R square on training data
regression_model.score(X_train, y_train)
```

```python
# R square on testing data
regression_model.score(X_test, y_test)
```

```python
# Import the metrics library
from sklearn import metrics
```

```python
#RMSE on Training data
predicted_train=regression_model.fit(X_train, y_train).predict(X_train)
np.sqrt(metrics.mean_squared_error(y_train,predicted_train))
```

```python
#RMSE on Testing data
predicted_test=regression_model.fit(X_train, y_train).predict(X_test)
np.sqrt(metrics.mean_squared_error(y_test,predicted_test))
```

## Insights:
    * Yes outliners has been detected in all items **These outliers value needs to be treated and there are several ways of treating them:
    * Drop the outlier value
    *Replace the outlier value using the IQR

```python
# Define a function which results the upper and lower limit to detect outliers for each fields
def remove_outlier(col):
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range   
```

```python
# Cap & floor the values beyond the outlier boundaries
for i in feature_list:
    LL,UL = remove_outlier(cmd[i])
    cmd[i] = np.where (cmd[i] > UL,UL, cmd[i])
    cmd[i] = np.where (cmd[i] < LL,LL, cmd[i])
```
